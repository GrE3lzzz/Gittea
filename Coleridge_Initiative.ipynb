```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive, files

# Mount Google Drive and upload the dataset
drive.mount('/content/drive')
uploaded = files.upload()

# Load the dataset
file_path = 'additional_gov_datasets_22000popular.csv'  # Update with the correct file name if different
dataset = pd.read_csv(file_path)
```

```python
# Basic statistics
print("\nBasic Statistics:")
print(dataset.info()) # Changed 'data' to 'dataset'
print(dataset.describe()) # Changed 'data' to 'dataset'
```

```python
# Text length distribution
dataset['text_length'] = dataset['text'].astype(str).apply(len) # Changed to convert to string type before applying len()
plt.figure(figsize=(10, 6))
sns.histplot(dataset['text_length'], bins=50, kde=True) # Changed 'data' to 'dataset'
plt.title("Distribution of Text Lengths")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Dataset mention frequency
from collections import Counter

# Print the available columns to check the actual name
print(dataset.columns)

# Since the 'title' column contains strings, we'll split them into words and count those
all_mentions = [word for title in dataset['title'] for word in title.lower().split()]
mention_counts = Counter(all_mentions)

print("\nTop 10 Dataset Mentions:")
print(mention_counts.most_common(10))
```

```python
# Visualize top dataset mentions
top_mentions = dict(mention_counts.most_common(10))
plt.figure(figsize=(10, 6))
sns.barplot(x=list(top_mentions.values()), y=list(top_mentions.keys()))
plt.title("Top 10 Most Frequent Dataset Mentions")
plt.xlabel("Frequency")
plt.ylabel("Dataset Mention")
plt.show()
```
